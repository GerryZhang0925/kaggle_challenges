```python
df_try = df[df['horsepower']!='?'].copy()
df_try['horsepower'] = df_try['horsepower'].astype(int)
plt.scatter(df_try['horsepower'], df_try['mpg'])
```

```python
def get_first_name(carname):
    for i, c in enumerate(carname):
        if (c == ' '):
            return carname[:i]
        
car_list = []
for car in df_try['car name']:
    car_list.append(get_first_name(car))

df_try['short name'] = car_list
df_try['short name'].value_counts()
sns.boxplot(x='short name', y='mpg', data=df_try)
```
